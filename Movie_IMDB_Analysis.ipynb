```python
# Find Movie Based on Ratings [Excellent, Good and Average ]
df.columns
```

```python
def rating(rating):
  if rating>=7.0:
    return 'Excellent'
  elif rating>6.0:
      return 'good'

  else:
        return 'Average'
```

```python
df['Rating_catagory']=df['Rating'].apply(rating)
```
